```python
import pandas as pd 
import os
```

```python
archivo = open('API_14_DS2_es_csv_v2_4533287.csv','r') # Abrimos archivo
# Separacion de datos
data = []
for fila in archivo:
    dato = fila.split(',"')     # Elemento a detectar para separacion de datos          
    data.append(dato)
archivo.close()


# Limpieza de datos 
rep = []
for i in range(0,len(data)):
    fila = data[i]
    fil = []
    for k in fila:
        dato = k.strip('"')
        fil.append(dato)
    rep.append(fil)

```

```python


# Ingesta de datos a Dataframe 
data_db = pd.DataFrame(columns = rep[0])
for colum in range(0,len(data_db.columns)):
    data_db[f'{data_db.columns[colum]}'] = pd.Series([i[colum] for i in rep])
```

```python

data_db.drop(0,inplace = True)  # Borramos la primera fila
data_db.drop(columns=['\n','Indicator Name'],inplace = True) # Eliminamos la ultima columna
años = ['Country Code', 'Indicator Code','1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997','1998', '1999', 
        '2000', '2001', '2002', '2003', '2004', '2005', '2006','2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
        '2016', '2017', '2018', '2019', '2020', '2021']
        
data_db = data_db[años] # Columnas desde 1990 - 2021
data_db = data_db.replace({"": 0})      # reemplazamos valores vacios con 0

for año in range(1990,2022):
    data_db[f'{año}'] = data_db[f'{año}'].astype('float')       # Cambio de tipo de dato (float)
```

```python
data_db[data_db['Country Code'] == 'FRA']
```

```python
# --------- REORDENAMIENTO DE DATOS
# ESTRUCTURA:

# Country Code | Año | Valor | Indicator Code
#       ...    | ... | ....  |  ........
# -------------|-----|-------|---------------

# Elementos para filtrado de datos 
years = list(range(1990,2021))
paises = data_db['Country Code'].unique()
indicator_code = data_db['Indicator Code'].unique()

# Creamos un dataframe como plantilla para la ingesta de los datos 
plantilla = {'Country Code':[],
           'Año':[],
           'Valor':[],
           'Indicator Code':[]}
data_mod = pd.DataFrame(plantilla)      # Dataframe para ingesta

# Filtrado y concatenado de los datos al nuevo modelo de tabla 
for pais in paises:
    data_pais = data_db[data_db['Country Code'] == pais]
    data_pais
    for indic in indicator_code:
        data_ind = data_pais[data_pais['Indicator Code'] == indic]
        data_años = data_ind[[str(año) for año in range(1990,2022)]]
    
        plantilla = {'Country Code':[pais]*32,
            'Año':[int(año) for año in range(1990,2022)],
            'Valor':list(data_años.iloc[0]),
            'Indicator Code':[indic]*32}
        data_mod_ind = pd.DataFrame(plantilla)
        data_mod = pd.concat([data_mod, data_mod_ind])      # Concatenado de datos 

data_mod['Año'] = data_mod['Año'].astype(int)       # Cambio del tipo de dato
```

```python
data_mod.to_csv('Data_new_table.csv',index = False,sep= ';')
```
